```python
from kaggle.api.kaggle_api_extended import KaggleApi

# Initialize and authenticate
api = KaggleApi()
api.authenticate()
print("âœ“ Kaggle API authenticated successfully!")
```

```python
dataset_name = 'abdelrhmanragab/european-soccer-database'
download_path = r'D:\Intern\AWS Lambda\Dataset'

```

```python
Path(download_path).mkdir(parents=True, exist_ok=True)

print(f"\nðŸ“¥ Downloading dataset: {dataset_name}")
print(f"ðŸ“‚ Download location: {download_path}")
```

```python
api.dataset_download_files(
    dataset_name,
    path=download_path,
    unzip=True  # Automatically extract files
)
```

```python
downloaded_files = list(Path(download_path).iterdir())
print(f"\nðŸ“„ Downloaded {len(downloaded_files)} file(s):")
for i, file in enumerate(downloaded_files, 1):
    size_mb = file.stat().st_size / (1024 * 1024)
    print(f"  {i}. {file.name} ({size_mb:.2f} MB)")

# Step 6: List CSV files specifically
csv_files = list(Path(download_path).glob('*.csv'))
print(f"\nðŸ“Š CSV files: {len(csv_files)}")
for csv in csv_files:
    print(f"  â€¢ {csv.name}")
```
