```python

from helpers import short_sleep, human_sleep

# File to store cookies
COOKIE_FILE = "instagram_cookies.pkl"

def load_cookies(driver, cookie_file):
    """Load cookies from file into the browser"""
    if os.path.exists(cookie_file):
        with open(cookie_file, "rb") as file:
            cookies = pickle.load(file)
            for cookie in cookies:
                driver.add_cookie(cookie)
        driver.refresh()
        short_sleep()
        return True
    return False

def save_cookies(driver, cookie_file):
    """Save cookies from browser to file"""
    cookies = driver.get_cookies()
    with open(cookie_file, "wb") as file:
        pickle.dump(cookies, file)

driver = webdriver.Chrome()
short_sleep()

# Open the webpage
driver.get("http://www.instagram.com")
short_sleep()

# Try to load cookies
if load_cookies(driver, COOKIE_FILE):
    print("Logged in using saved cookies")
else:
    print("No cookies found, performing manual login")
    
    # Perform login
    username = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='username']"))
    )
    password = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='password']"))
    )
    short_sleep()
    
    username.clear()
    username.send_keys("your instagram username / gamil")
    password.clear()
    password.send_keys("your password")
    short_sleep()
    
    # Target the login button and click it
    button = WebDriverWait(driver, 2).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']"))
    ).click()
    
    short_sleep()
    
    # Handle NOT NOW
    try:
        not_now = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Not Now")]'))
        ).click()
    except:
        pass
    
    short_sleep()
    
    # Save cookies after successful login
    save_cookies(driver, COOKIE_FILE)
    print("Cookies saved for future use")

# Continue with your automation tasks here
print("Ready to proceed with Instagram automation")
```

```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os


human_sleep()

# Define the file path
file_path = r'C:\Users\Usman Asghar\Desktop\Intern\Instagram Web Scraping\Excel\profile_data.xlsx'

# Read the existing Excel file
df = pd.read_excel(file_path)



# Create list to store the combined bio data
bio_list = []

short_sleep()

# Iterate through each row in the dataframe
for index, row in df.iterrows():
    profile_url = row['Profile URL']
    full_name = row['Name']

    
    # Navigate directly to the Instagram profile
    driver.get(profile_url)
    
    print(f"\n[{index + 1}/{len(df)}] Navigating to: {profile_url}")
    print(f"   Full Name: {full_name}")
    
    
    human_sleep()

    try:
        # Wait for header section to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//header"))
        )
        short_sleep()

        # Initialize components
        bio_text = ""
        link_url = ""
        link_text = ""
        threads_username = ""
        
        # Extract Bio Text - Multiple methods
        try:
            # Method 1: Direct parent div approach
            bio_element = driver.find_element(By.XPATH, 
                "//div[contains(@class, 'x7a106z x972fbf')]//div[@role='button']//span[@dir='auto']")
            bio_text = bio_element.text.strip()
            print(f"  ‚úì Bio extracted (Method 1)")
        except:
            try:
                # Method 2: Target outer span
                bio_element = driver.find_element(By.XPATH, 
                    "//div[contains(@class, 'x7a106z x972fbf')]//span[@class='_ap3a _aaco _aacu _aacx _aad7 _aade'][@dir='auto']")
                bio_text = bio_element.text.strip()
                print(f"  ‚úì Bio extracted (Method 2)")
            except:
                try:
                    # Method 3: Any span with specific classes in header
                    bio_element = driver.find_element(By.XPATH, 
                        "//header//span[@class='_ap3a _aaco _aacu _aacx _aad7 _aade'][@dir='auto']")
                    bio_text = bio_element.text.strip()
                    print(f"  ‚úì Bio extracted (Method 3)")
                except:
                    try:
                        # Method 4: CSS Selector
                        bio_element = driver.find_element(By.CSS_SELECTOR, 
                            "div.x7a106z span._ap3a._aaco._aacu._aacx._aad7._aade[dir='auto']")
                        bio_text = bio_element.text.strip()
                        print(f"  ‚úì Bio extracted (Method 4)")
                    except:
                        print(f"  ‚ö†Ô∏è Bio not found")
                        bio_text = ""
        
        # Extract Link URL and Link Text
        try:
            # Find link with rel attribute
            link_element = driver.find_element(By.XPATH, 
                "//div[contains(@class, 'x7a106z x972fbf')]//a[@rel='me nofollow noopener noreferrer']")
            link_url = link_element.get_attribute('href')
            
            try:
                # Extract display text
                link_text_element = driver.find_element(By.XPATH, 
                    "//div[contains(@class, 'x7a106z x972fbf')]//a[@rel='me nofollow noopener noreferrer']//div[@dir='auto']")
                link_text = link_text_element.text.strip()
                print(f"  ‚úì Link extracted: {link_text}")
            except:
                link_text = link_url
                print(f"  ‚úì Link URL extracted")
        except:
            
                print(f"  ‚ö†Ô∏è Link not found")
                pass
        
        # Extract Threads Username
        try:
            # Method 1: Direct threads link
            threads_link = driver.find_element(By.XPATH, 
                "//div[contains(@class, 'x7a106z x972fbf')]//a[contains(@href, 'threads.com')]")
            
            try:
                # Extract username from span
                threads_username_element = driver.find_element(By.XPATH, 
                    "//div[contains(@class, 'x7a106z x972fbf')]//a[contains(@href, 'threads.com')]//span[contains(@class, 'x1lliihq x193iq5w')]")
                threads_username = threads_username_element.text.strip()
                print(f"  ‚úì Threads: @{threads_username}")
            except:
                # Extract from href
                threads_href = threads_link.get_attribute('href')
                if '@' in threads_href:
                    threads_username = threads_href.split('@')[1].split('?')[0]
                    print(f"  ‚úì Threads: @{threads_username} (from URL)")
        except:
            try:
                # Alternative: Any threads link
                threads_link = driver.find_element(By.XPATH, 
                    "//header//a[contains(@href, 'threads.com')]")
                threads_href = threads_link.get_attribute('href')
                if '@' in threads_href:
                    threads_username = threads_href.split('@')[1].split('?')[0]
                    print(f"  ‚úì Threads: @{threads_username} (alternative)")
            except:
                print(f"  ‚ö†Ô∏è Threads not found")
                pass
        
        # Combine all information into one string with N/A for missing data
        combined_bio = []
        
        # Add bio text or N/A
        combined_bio.append(f"Bio: {bio_text if bio_text else 'N/A'}")
        
        # Add link text or N/A
        combined_bio.append(f"Link: {link_text if link_text else 'N/A'}")
        

        
        # Add threads username or N/A
        combined_bio.append(f"Threads: @{threads_username if threads_username else 'N/A'}")
        
        # Join all parts with separator
        final_bio = " | ".join(combined_bio)
        
        print(f"  üìù Combined Bio: {final_bio[:150]}..." if len(final_bio) > 150 else f"  üìù Combined Bio: {final_bio}")
        bio_list.append(final_bio)
        
    except Exception as e:
        print(f"‚úó Failed to load profile: {str(e)}")
        bio_list.append("N/A")
    
    # Update dataframe with current bio data
    df.loc[:len(bio_list)-1, 'Bio'] = bio_list
    
    # Save to Excel immediately after extracting this profile
    df.to_excel(file_path, index=False)
    print(f"üíæ ‚úì Profile {index + 1}/{len(df)} saved to Excel!")
    print("-" * 60)
    
    short_sleep()

print("\n" + "="*60)
print("‚úì‚úì‚úì EXTRACTION COMPLETE! ‚úì‚úì‚úì")
print(f"Total profiles processed: {len(df)}")
print(f"Results saved to: {file_path}")
print("="*60)
print("\nFinal Results:")
```

```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os

def format_number(num_str):
    """Convert number string to K/M/B format"""
    try:
        # Remove commas and convert to integer
        num = int(num_str.replace(',', '').strip())
        
        if num >= 1_000_000_000:  # Billions
            return f"{num / 1_000_000_000:.1f}B"
        elif num >= 1_000_000:  # Millions
            return f"{num / 1_000_000:.1f}M"
        elif num >= 1_000:  # Thousands
            return f"{num / 1_000:.1f}K"
        else:
            return str(num)
    except:
        return num_str

human_sleep()

# Define the file path
file_path = r'C:\Users\Usman Asghar\Desktop\Intern\Instagram Web Scraping\Excel\profile_data.xlsx'

# Read the existing Excel file
df = pd.read_excel(file_path)

driver.maximize_window()

# Create lists to store the new data
posts_list = []
followers_list = []
following_list = []


short_sleep()

# Iterate through each row in the dataframe
for index, row in df.iterrows():
    profile_url = row['Profile URL']
    full_name = row['Name']
    
    
    # Navigate directly to the Instagram profile
    driver.get(profile_url)
    
    print(f"\n[{index + 1}/{len(df)}] Navigating to: {profile_url}")
    print(f"   Full Name: {full_name}")
 
    
    human_sleep()

    try:
        # Wait for stats to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//header"))
        )
        short_sleep()
        
        # Extract posts, followers, and following
        try:
            # Method 1: Get all stat numbers
            stats_elements = driver.find_elements(By.XPATH, "//header//span[@class='html-span xdj266r x14z9mp xat24cr x1lziwak xexx8yu xyri2b x18d9i69 x1c1uobl x1hl2dhg x16tdsg8 x1vvkbs']")
            
            if len(stats_elements) >= 3:
                posts_raw = stats_elements[0].text.replace(',', '').strip()
                
                # For followers, try to get from title attribute first (shows full number)
                try:
                    followers_element = driver.find_element(By.XPATH, "//a[contains(@href, '/followers/')]//span[@title]")
                    followers_raw = followers_element.get_attribute('title').replace(',', '').strip()
                except:
                    followers_raw = stats_elements[1].text.replace(',', '').strip()
                
                following_raw = stats_elements[2].text.replace(',', '').strip()
                
                # Format numbers to K/M/B
                posts = format_number(posts_raw)
                followers = format_number(followers_raw)
                following = format_number(following_raw)
            else:
                raise Exception("Stats elements not found with Method 1")
            
            print(f"  üìä Posts: {posts}")
            print(f"  üë• Followers: {followers}")
            print(f"  üë§ Following: {following}")
            
            # Append extracted data to lists
            posts_list.append(posts)
            followers_list.append(followers)
            following_list.append(following)
            
        except Exception as stat_error:
            print(f"‚ö†Ô∏è Could not extract stats: {str(stat_error)}")
            posts_list.append("N/A")
            followers_list.append("N/A")
            following_list.append("N/A")
        
        
        
    except Exception as e:
        print(f"‚úó Failed to load profile: {str(e)}")
        posts_list.append("N/A")
        followers_list.append("N/A")
        following_list.append("N/A")
   
    
    # ‚≠ê SAVE AFTER EACH PROFILE IS EXTRACTED ‚≠ê
    # Update the dataframe with the current extracted data
    df.loc[:len(posts_list)-1, 'Posts'] = posts_list
    df.loc[:len(followers_list)-1, 'Followers'] = followers_list
    df.loc[:len(following_list)-1, 'Following'] = following_list
  
    
    # Save to Excel immediately after extracting this profile
    df.to_excel(file_path, index=False)
    print(f"üíæ ‚úì Profile {index + 1}/{len(df)} saved to Excel!")
    print("-" * 60)
    
    short_sleep()

print("\n" + "="*60)
print("‚úì‚úì‚úì EXTRACTION COMPLETE! ‚úì‚úì‚úì")
print(f"Total profiles processed: {len(df)}")
print(f"Results saved to: {file_path}")
print("="*60)
print("\nFinal Results:")
print(df[['Name', 'Posts', 'Followers', 'Following']])
```

```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

human_sleep()

# Define the file path
file_path = r'C:\Users\Usman Asghar\Desktop\Intern\Instagram Web Scraping\Excel\profile_data.xlsx'

# Read the existing Excel file
df = pd.read_excel(file_path)



# Create lists to store comments (5 separate columns)
comment_1_list = []
comment_2_list = []
comment_3_list = []
comment_4_list = []
comment_5_list = []

short_sleep()

# Iterate through each row in the dataframe
for index, row in df.iterrows():
    profile_url = row['Profile URL']
    full_name = row['Name']

    
    # Navigate directly to the Instagram profile
    driver.get(profile_url)
    
    print(f"\n[{index + 1}/{len(df)}] Navigating to: {profile_url}")
    print(f"   Full Name: {full_name}")

    
    human_sleep()

    # Initialize comment variables
    comments = ["N/A", "N/A", "N/A", "N/A", "N/A"]
    
    try:
        # Wait for the first post to be clickable
        first_post = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//a[contains(@href, '/p/')]"))
        )
        
        print(f"  ‚úì First post found, clicking...")
        first_post.click()
        
        short_sleep()
        
        # Wait for the post modal to open and comments to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//ul[contains(@class, '_a9z6')]"))
        )
        
        print(f"  ‚úì Post opened, extracting comments...")
        
        # Extract comments using multiple methods
        try:
            # Method 1: Target comment text within list items
            comment_elements = driver.find_elements(By.XPATH, 
                "//ul[contains(@class, '_a9z6')]//li[contains(@class, '_a9zj')]//span[@class='_ap3a _aaco _aacu _aacx _aad7 _aade'][@dir='auto']")
            
            # Extract up to 5 comments
            for i, element in enumerate(comment_elements[:5]):
                try:
                    comment_text = element.text.strip()
                    if comment_text:
                        comments[i] = comment_text
                        print(f"  ‚úì Comment {i+1}: {comment_text[:50]}..." if len(comment_text) > 50 else f"  ‚úì Comment {i+1}: {comment_text}")
                except:
                    continue
            
            if comments[0] == "N/A":
                print(f"  ‚ö†Ô∏è No comments found (Method 1)")
                
                # Method 2: Alternative XPath
                try:
                    comment_elements = driver.find_elements(By.XPATH, 
                        "//div[contains(@class, '_a9zr')]//span[@dir='auto']")
                    
                    for i, element in enumerate(comment_elements[:5]):
                        try:
                            comment_text = element.text.strip()
                            if comment_text and len(comment_text) > 3:  # Filter out very short texts
                                comments[i] = comment_text
                                print(f"  ‚úì Comment {i+1}: {comment_text[:50]}..." if len(comment_text) > 50 else f"  ‚úì Comment {i+1}: {comment_text}")
                        except:
                            continue
                except:
                    print(f"  ‚ö†Ô∏è No comments found (Method 2)")
        
        except Exception as e:
            print(f"  ‚úó Failed to extract comments: {str(e)}")
        
        # Close the post modal by pressing ESC or clicking back
        try:
            driver.execute_script("window.history.go(-1)")
            short_sleep()
        except:
            pass
            
    except Exception as e:
        print(f"  ‚úó Failed to open post: {str(e)}")
    
    # Append comments to lists
    comment_1_list.append(comments[0])
    comment_2_list.append(comments[1])
    comment_3_list.append(comments[2])
    comment_4_list.append(comments[3])
    comment_5_list.append(comments[4])
    
    # Update dataframe with current comment data
    df.loc[:len(comment_1_list)-1, 'Comment 1'] = comment_1_list
    df.loc[:len(comment_2_list)-1, 'Comment 2'] = comment_2_list
    df.loc[:len(comment_3_list)-1, 'Comment 3'] = comment_3_list
    df.loc[:len(comment_4_list)-1, 'Comment 4'] = comment_4_list
    df.loc[:len(comment_5_list)-1, 'Comment 5'] = comment_5_list
    
    # Save to Excel immediately after extracting comments from this profile
    df.to_excel(file_path, index=False)
    print(f"üíæ ‚úì Profile {index + 1}/{len(df)} saved to Excel!")
    print("-" * 60)
    
    short_sleep()

print("\n" + "="*60)
print("‚úì‚úì‚úì COMMENT EXTRACTION COMPLETE! ‚úì‚úì‚úì")
print(f"Total profiles processed: {len(df)}")
print(f"Results saved to: {file_path}")
print("="*60)
print("\nFinal Results Preview:")
print(df[['Name', 'Comment 1', 'Comment 2', 'Comment 3']].head().to_string())
```

```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

human_sleep()

# Define the file path
file_path = r'C:\Users\Usman Asghar\Desktop\Intern\Instagram Web Scraping\Excel\profile_data.xlsx'

# Read the existing Excel file
df = pd.read_excel(file_path)

# Create list to store profile image URLs
profile_image_list = []

short_sleep()

# Iterate through each row in the dataframe
for index, row in df.iterrows():
    profile_url = row['Profile URL']
    full_name = row['Name']

    
    # Navigate directly to the Instagram profile
    driver.get(profile_url)
    
    print(f"\n[{index + 1}/{len(df)}] Navigating to: {profile_url}")
    print(f"   Full Name: {full_name}")
   
    
    human_sleep()

    # Initialize profile image URL
    profile_image_url = "N/A"
    
    try:
        # Wait for profile image to load and extract src
        profile_img_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//img[@alt and @crossorigin='anonymous']"))
        )
        
        # Get the src attribute from the img tag
        img_src = profile_img_element.get_attribute('src')
        
        if img_src and len(img_src) > 0:
            profile_image_url = img_src
            print(f"  ‚úì Profile image found: {img_src[:80]}...")
        else:
            print(f"  ‚ö†Ô∏è Profile image src is empty")
            
    except Exception as e:
        print(f"  ‚úó Failed to extract profile image: {str(e)}")
        
        # Try alternative method - look for profile picture by class
        try:
            alternative_img = driver.find_element(By.XPATH, 
                "//img[contains(@class, 'xpdipgo') or contains(@class, '_aa8i')]")
            img_src = alternative_img.get_attribute('src')
            
            if img_src and len(img_src) > 0:
                profile_image_url = img_src
                print(f"  ‚úì Profile image found (alternative method): {img_src[:80]}...")
            else:
                print(f"  ‚ö†Ô∏è No profile image available (Method 2)")
        except:
            print(f"  ‚ö†Ô∏è No profile image available (both methods failed)")
    
    # Append profile image URL to list
    profile_image_list.append(profile_image_url)
    
    # Update dataframe with current profile image data
    df.loc[:len(profile_image_list)-1, 'Profile Image URL'] = profile_image_list
    
    # Save to Excel immediately after extracting image from this profile
    df.to_excel(file_path, index=False)
    print(f"üíæ ‚úì Profile {index + 1}/{len(df)} saved to Excel!")
    print("-" * 60)
    
    short_sleep()

print("\n" + "="*60)
print("‚úì‚úì‚úì PROFILE IMAGE EXTRACTION COMPLETE! ‚úì‚úì‚úì")
print(f"Total profiles processed: {len(df)}")
print(f"Results saved to: {file_path}")
print("="*60)
```
